```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
image = plt.imread('butterFly.gif')
def show(img):
    plt.imshow(img, cmap='gray')
    plt.show()
```

2(a). Brightness enhancement of a specific range of gray level


```python

# Define the range of gray levels to enhance (adjust these values as needed)
min_gray_level = int(input("Enter min gray level: "))
max_gray_level = int(input("Enter max gray level: "))

# Define the enhancement factor (adjust this as needed)
enhancement_factor = 1.5  # You can change this value

# Apply the brightness enhancement
enhanced_image = image.copy()
for x in range(image.shape[0]):
    for y in range(image.shape[1]):
        pixel_value = image[x, y]
        if min_gray_level <= pixel_value <= max_gray_level:
            new_pixel_value = pixel_value * enhancement_factor
            if new_pixel_value > 255:
                new_pixel_value = 255
            enhanced_image[x, y] = new_pixel_value

# Display the original and enhanced images
plt.imshow(image, cmap='gray')
plt.title('original image')
plt.show()

plt.imshow(enhanced_image, cmap='gray')
plt.title('enhanced image')
plt.show()
```

2(b). Power Law transform :

```python
def law_transform(img):
    law_transform_img = img.copy()
    for gamma in [0.1, 1.0, 2.0, 4.5]:
        law_transform_img =  255 * ((img / 255) ** gamma)
        law_transform_img = law_transform_img.astype(np.uint8)
        plt.imshow(law_transform_img, cmap = 'gray')
        plt.show()
```

```python
law_transform(image)
```

Inverse log transform :

```python
def in_log_transform(img):
    in_img = img.copy()
    c = 255 // np.log(1 + np.max(img))
    in_img = np.exp(img / c) - 1
    plt.imshow(in_img, cmap = 'gray')
    plt.show()
```

```python
in_log_transform(image)
plt.imshow(image, cmap='gray')
plt.show()
```

2(c). Difference between original image and image obtained by last three bits(MSB):

```python
def last_three_bit(img):
    mask = 0b11100000
    new_img = img & mask
    return new_img
```

```python
img1 = last_three_bit(image)
diff = image - img1
# show(image)
# show(img1)
# show(diff)
plt.subplot(1 , 3 , 1)
plt.imshow(image, cmap='gray')
# plt.show()
plt.subplot(1 , 3 , 2)
plt.imshow(img1, cmap='gray')
plt.subplot(1 , 3 , 3)
# plt.show()
plt.imshow(diff, cmap='gray')
plt.show()
```
