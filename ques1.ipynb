1(a): Decrease spatial resolution by half every time 

```python
import numpy as np
```

```python
def down_sample(img, F):
    f = F
    m , n = img.shape

    img1 = np.zeros((m // f, n // f), dtype = np.uint8)

    for i in range(0, m):
        for j in range(0, n):
                img1[i // f][j // f] = img[i][j]
            
            
    plt.imshow(img1, cmap = 'gray')
    plt.show()
```

```python
F = 1
while F <= 512 :
    down_sample(image, F)
    F *= 2
```

1(b) :
decrease intensity level resulation by one bit up to reach binary format :


```python
def intensity_level(b, img):
    m, n = img.shape
    img1 = img.copy()
    for i in range(0, m):
        for j in range(0, n):
            img1[i][j] = (img[i][j] // 2 ** b) * 2 ** b

    plt.imshow(img1, cmap = 'gray')
    plt.show()
```

```python
b = 0
while b <= 7 :
    intensity_level(b, image)
    b += 1
```

1(c) : Illustrate the histogram of the image

```python
def histogram(img) :
    hist = np.zeros(256, dtype = np.uint8)
    m, n = img.shape
    for i in range(0, m) :
        for j in range(0, n) :
            hist[img[i, j]] += 1
    plt.bar(range(0, 256), hist)
    plt.xlabel("intensity ")
    plt.ylabel("frequencies ")
    plt.title("histogram")
    # plt.plot(hist)
    plt.show()
```

```python
histogram(image)
```

Make single thresold segmentation observed from the histogram

```python
def single_thresold_segmentation(img):
    single_thresold_segmentation_img = img.copy()
    m, n = img.shape
    thresold = np.average(img)
    #thresold = 128
    for i in range(0, m):
        for j in range(0, n):
            if img[i][j] > thresold :
                single_thresold_segmentation_img[i][j] = 255
            else:
                single_thresold_segmentation_img[i][j] = 0
    plt.imshow(single_thresold_segmentation_img, cmap = 'gray')
    plt.show()
```

```python
single_thresold_segmentation(image)
```
